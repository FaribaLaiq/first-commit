```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
#read the seleced dataset of ML
df = pd.read_csv('/kaggle/input/edgeiiotset-cyber-security-dataset-of-iot-iiot/Edge-IIoTset dataset/Selected dataset for ML and DL/ML-EdgeIIoT-dataset.csv', low_memory=False)

#Filter DDOS and Normal traffic data from all data
df=df.loc[df['Attack_type'].isin(['DDoS_HTTP', 'DDoS_ICMP', 'DDoS_TCP', 'DDoS_UDP', 'Normal'])]

#display df
display(df.head(5))

#some basic statistics
print(df['Attack_type'].value_counts())
print(len(df.index))
df.describe()
```

# Drop Unnecessary columns and remove duplicates

```python
from sklearn.utils import shuffle

#drop unecessary columns (done by the previous author)
drop_columns = ["frame.time", "ip.src_host", "ip.dst_host", "arp.src.proto_ipv4","arp.dst.proto_ipv4", 

         "http.file_data","http.request.full_uri","icmp.transmit_timestamp",

         "http.request.uri.query", "tcp.options","tcp.payload","tcp.srcport",

         "tcp.dstport", "udp.port", "mqtt.msg"]

df.drop(drop_columns, axis=1, inplace=True)

df.dropna(axis=0, how='any', inplace=True)

#remove duplicates
df.drop_duplicates(subset=None, keep="first", inplace=True)

#shuffle records
df = shuffle(df)

#check missing values
print(df.isna().sum())

#print number of records
print(df['Attack_type'].value_counts())
print("Total records: ", len(df.index))
print("Total columns after removing unnecessary columns: ", len(df.columns))
df.head(5)
```

# Label encoding

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Iterate over column names
for column in df:
    if df[column].dtype.kind == 'O':
        print(df[column])
        df[column]= label_encoder.fit_transform(df[column])
        df[column]= label_encoder.fit_transform(df[column]) 
        
display(df.head(20))

```

# Apply SMOTE

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Assuming your dataset is stored in a pandas DataFrame called 'df'
X, y = df.iloc[:,:-2],df.iloc[:,-1]

seed = 42
test_size = 0.30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)

class_counts = y_train.value_counts()
print(class_counts)


class_counts = y_test.value_counts()
print(class_counts)
# Creating an instance of the SMOTE oversampler
smote = SMOTE()

# Resampling the dataset using SMOTE
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Creating a new balanced DataFrame
df_balanced = pd.concat([pd.DataFrame(X_resampled), pd.DataFrame(y_resampled)], axis=1)

# Checking the class distribution in the balanced dataset
class_counts = df_balanced['Attack_type'].value_counts()
print(class_counts)
```
